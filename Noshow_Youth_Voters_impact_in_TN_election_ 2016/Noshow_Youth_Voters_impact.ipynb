# **Problem Statement**
 It is said that 1 in 5 youth (i.e. 20%) is not interested in voting, there are multiple reasons behind this. But what is the impact of these voters in election?. Let us take 2016 Tamilnadu assembly election result and find their impact.

```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

**poll_data**  --> 2016 TN election result taken from Lok dhaba 

**census_data** --> 2011 Distict Wise youth population data taken govt census report.

```
poll_data = pd.read_csv('TCPD_AE_Tamil Nadu_2020-6-25.csv')
```

```
poll_data.head()
```

```
census_data = pd.read_excel('DDW-3300C-14.xls')
```

# **Data Cleaning**

```
census_data.head(10)
```

```
census_data=census_data.iloc[5:]
```

```
census_data.head()
```

# **Data Processing**

```
census_data.columns
```

```
census_data = census_data[['Distt.', 'Area Name', 'Age-group', 'Total ']]
```

```
census_data.reset_index(drop=True)
```

```
youth_data=census_data[(census_data['Age-group'] == '10-14')|(census_data['Age-group'] == '15-19')]
```

```
youth_data.head()
```

```
youth_data = youth_data.groupby(by=['Distt.','Area Name']).sum()
```

```
youth_data.to_csv('youth.csv') 
```

```
youth_final = pd.read_csv('youth.csv')
```

```
youth_final.tail()
```

```
poll_data.head()
```

```
poll_data = poll_data[poll_data['Year']==2016]
```

```
poll_data
```

```
poll_data.columns
```

```
poll_data = poll_data[['Constituency_No','Position','Party','Margin','District_Name']]
```

```
poll_data
```

```
poll_data = poll_data[(poll_data['Position'] == 1)|(poll_data['Position'] == 2)]
```

```
poll_data['District_Name'].iloc[0]
```

```
poll_data
```

```
poll_data['District_Name'].nunique()
```

```
youth_final.drop('Distt.',axis=1,inplace=True)
```

```
youth_final.head()
```

```
def youth_count(dist):
  #print(youth_final[youth_final['Area Name']==dist]['Total '].iloc[0])
  return youth_final.loc[youth_final['Area Name']==dist]['Total '].iloc[0]
#youth_count(poll_data['District_Name'].iloc[0]) 
```

```
poll_data['youth'] = poll_data['District_Name'].apply(youth_count)
```

```
poll_data['dist_ac']= poll_data['District_Name'].apply( lambda x: poll_data[poll_data['District_Name']==x]['Constituency_No'].nunique())
```

```
poll_data.head()
```

```
poll_data['ac_youth_avg'] = poll_data['youth'] / poll_data['dist_ac']
```

```
poll_data
```

```
poll_data['noshow_youth'] = poll_data['ac_youth_avg'] / 100 *20
```

```
poll_data.head()
```

```
poll_data[poll_data['Position']==1]['Party'].value_counts()
```

```
def diff(ac_no):
   return poll_data[(poll_data['Constituency_No'] == ac_no) & (poll_data['Position'] == 1)]['Margin'].iloc[0]
#diff(1)   
```

```
poll_data['diff'] = poll_data['Constituency_No'].apply(diff)
```

```
poll_data.head()
```

```
poll_data['25%'] = poll_data['noshow_youth'] / 100 * 25
```

```
poll_data['50%'] = poll_data['noshow_youth'] / 100 * 50
```

# **Exploratory Data Analysis**

## Admk could have won 13 seats more in 2016 election if they managed to get 25% vote of youth who were not interested in voting.(you can find the constituency no. in below dataframe)

```
poll_data[(poll_data['Position']==2) & (poll_data['Party']=='ADMK') & (poll_data['diff'] <= poll_data['25%'])]
```

Admk could have lost 17 seats if a party that is trailing ADMK managed to get hold of 25% of youth who were not interested in voting.(you can find the constituency no. in below dataframe)  

```
poll_data[(poll_data['Position']==1) & (poll_data['Party']=='ADMK') & (poll_data['diff'] <= poll_data['25%'])]
```

## Admk could have won 24 seats more in 2016 election if they managed to get 50% vote of youth who were not interested in voting.(you can find the constituency no. in below dataframe)

```
poll_data[(poll_data['Position']==2) & (poll_data['Party']=='ADMK') & (poll_data['diff'] <= poll_data['50%'])]
```

Admk could have lost 30 seats if a party that is trailing ADMK managed to get hold of 50% of youth who were not interested in voting.(you can find the constituency no. in below dataframe) 

```
poll_data[(poll_data['Position']==1) & (poll_data['Party']=='ADMK') & (poll_data['diff'] <= poll_data['50%'])]
```
