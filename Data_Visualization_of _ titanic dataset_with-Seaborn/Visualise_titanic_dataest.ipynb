## The Data

We will be working with a famous titanic data set for these exercises.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.jointplot(x='fare',y='age',data=titanic)
```

```python
sns.distplot(titanic['fare'],bins=30,kde=False,color='red')
```

```python
sns.boxplot(x='class',y='age',data=titanic,palette='rainbow')
```

```python
sns.swarmplot(x='class',y='age',data=titanic,palette='Set1',hue='alive')
```

```python
g=sns.violinplot(x='class',y='age',data=titanic,palette='Set1',hue='alive',split=True,legend_out = True)
g.legend(loc='center left', bbox_to_anchor=(1, 0.9), ncol=1)
```

```python
sns.countplot(x='sex',data=titanic,hue='alive')
```

```python
sns.heatmap(titanic.corr(),cmap='coolwarm')
plt.title('titanic.corr()')
```

```python
g = sns.FacetGrid(data=titanic,col='sex')
g.map(plt.hist,'age')
```
